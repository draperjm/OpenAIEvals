```python
pryn = {"цьмяны": ["бляклы", "матавы"],
 "зграбны": ["стройны", "статны", "грацыёзны"],
 "цікаўны": ["цікавы", "дапытлівы"],
 "лекавы": ["гаючы"],
 "дрэнны": ["благі", "кепскі", "паганы"]}
```

```python
list(itertools.product(pryn.keys(), itertools.chain.from_iterable(pryn.values())))
```

```python
answers = []
for word, syn in list(itertools.product(pryn.keys(), itertools.chain.from_iterable(pryn.values()))):
    if syn in pryn.get(word):
        answers.append((word, syn, 'Y'))
    else:
        answers.append((word, syn, 'N'))
```
